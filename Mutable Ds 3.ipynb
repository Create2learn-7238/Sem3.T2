# Lambda Function
- function without name
- nameless function is called anonymous function or lambda function

```python
# user define named function
def square(n):
    return n*n

print(square(5))
print(square(2))
```

```python
# lambda function
s=lambda n:n*n

print(s(5))
```

```python
d=lambda a,b:(a+b , a-b , a*b , a/b)
        
print(d(2,5))
print(d(5,2))
```

```python
s=lambda x:x.upper()
print(s("Dixit Patel"))
```

```python
words=['apple','banana','cherry','d','kiwi']
s=sorted(words,key=lambda x:len(x))
print(s)
```

```python
a=[(1,3),(4,2),(5,8),(9,1)]
a.sort(key=lambda x:x[1])
# s=sorted(a,key=lambda x:x[1])
print(a)
```

## 1) Map()
- for every element present in given sequence apply some functionality and generate new elemenet with req modification
- map(function,sequence)

```python
l=[1,2,3,4,5]
def double(n):
    return 2*n
l1=list(map(double,l))
print(l1)
```

```python
l=[6,7,8,9,10]
l1=list(map(lambda x:x*2,l))
print(l1)
```

```python
a=[1,2,3,4,5]
b=[2,3,4,5,6]
ans=list(map(lambda x,y:x*y ,a,b))
print(ans)
```

## reduce()
- reduce sequence of element into a single element by applying specific function
- reduce(function,sequence)

```python
from functools import *
l=[1,2,3,4]
sum=reduce(lambda x,y:x+y ,l)
print(sum)

m=reduce(lambda x,y:x*y ,l)
print(m)

sb=reduce(lambda x,y:x-y ,l)
print(sb)
```

## filter()
- used to filter the value from given sequence based on some condition
- filter(function,sequence)

```python
def even(x):
    if x%2==0:
        return True
    else:
        return False
    
l=[0,5,10,20,25,30]
l1=list(filter(even , l))
print(l1)
```

```python
l=[45,1,15,13,15,24,62,68,654]
l1=list(filter(lambda x:x%2==0,l))
print(l1)
```

```python
c=lambda x:"Eveeen" if x%2==0 else "Oddd"
print(c(5))
print(c(10))
```

## WAP to find sum of positive and negative number from the given list of number using lambda function

```python
l=[2,4,-6,-9,11,-12,14,-5,17]
p=lambda x:sum(n for n in l if n>0)
n=lambda x:sum(n for n in l if n<0)
print(p(l))
print(n(l))
```

```python
l=[2,4,-6,-9,11,-12,14,-5,17]
p=sum(filter(lambda x:x>0,l))
n=sum(filter(lambda x:x<0,l))
print(p)
print(n)
```
